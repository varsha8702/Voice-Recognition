```python
import speech_recognition as sr

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening... Speak now.")
        recognizer.adjust_for_ambient_noise(source)  # Reduce background noise
        audio = recognizer.listen(source)  # Capture audio

        try:
            text = recognizer.recognize_google(audio)  # Convert speech to text
            print("You said:", text)
        except sr.UnknownValueError:
            print("Sorry, could not understand the audio.")
        except sr.RequestError:
            print("Could not request results, check your internet connection.")

if __name__ == "__main__":
    recognize_speech()
```

```python
import speech_recognition as sr
import webbrowser

def recognize_speech():
    recognizer = sr.Recognizer()
    
    with sr.Microphone() as source:
        print("Listening... Speak your search query.")
        recognizer.adjust_for_ambient_noise(source)  # Reduce background noise
        audio = recognizer.listen(source)  # Capture audio

        try:
            query = recognizer.recognize_google(audio)  # Convert speech to text
            print(f"Searching for: {query}")

            # Open Google with the search query
            url = f"https://www.google.com/search?q={query}"
            webbrowser.open(url)
            print("Opening Google...")

        except sr.UnknownValueError:
            print("Sorry, could not understand the audio.")
        except sr.RequestError:
            print("Could not request results, check your internet connection.")

if __name__ == "__main__":
    recognize_speech()
```

```python
import cv2
import os
import subprocess
import webbrowser
import speech_recognition as sr

def open_camera():
    """Opens the webcam using OpenCV."""
    cap = cv2.VideoCapture(0)  # Open default camera
    if not cap.isOpened():
        print("Error: Could not open camera.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        cv2.imshow("Camera Feed", frame)

        # Press 'q' to close the camera
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

def open_camera_app():
    """Opens the built-in Camera app on Windows."""
    os.system("start microsoft.windows.camera:")

def open_application(app_name):
    """Opens common applications based on voice command."""
    apps = {
        "notepad": "notepad.exe",
        "calculator": "calc.exe",
        "chrome": "C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe",
        "file explorer": "explorer.exe",
        "camera": open_camera_app  # Function reference
    }

    if app_name in apps:
        if callable(apps[app_name]):  # If it's a function, call it
            apps[app_name]()
        else:
            subprocess.run(apps[app_name])
        print(f"Opening {app_name}...")
    else:
        print("Application not found!")

def search_google(query):
    """Searches Google using the voice-recognized query."""
    url = f"https://www.google.com/search?q={query}"
    webbrowser.open(url)
    print(f"Searching Google for: {query}")

def voice_command():
    """Recognizes speech and performs actions accordingly."""
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Listening for a command (say 'open camera', 'open notepad', 'search ...').")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

        try:
            command = recognizer.recognize_google(audio).lower()
            print(f"You said: {command}")

            if "open camera" in command:
                open_camera()
            elif "open" in command:
                app_name = command.replace("open ", "").strip()
                open_application(app_name)
            elif "search" in command:
                search_query = command.replace("search ", "").strip()
                search_google(search_query)
            else:
                print("Command not recognized.")

        except sr.UnknownValueError:
            print("Sorry, could not understand the audio.")
        except sr.RequestError:
            print("Could not request results, check your internet connection.")

if __name__ == "__main__":
    voice_command()
```
